```python
# a)
import matplotlib.pyplot as plt
import numpy as np

A = 1.024e-23 / (1.381e-23)
B = 1.582e-26 / (1.381e-23)
r = np.linspace(-2, 2, 100)

U = B / r**12 - A / r ** 6
F = (12 * B) / r ** 13 - (6 * A) / r ** 7

plt.ylim(-5, 5)

plt.plot(r, U)
plt.plot(r, F)
plt.show()

# Ok so, I was looking through the wikipedia page of the Lennard-Jones formula and it said that  = B^2 / 4A... that would give us a value of 2.4326066402741967e-07.
# Plugging the numbers into desmos, I got r_0 values of *maybe* 0.34 (process of elimination lol).
```

```python
# b)
A = 1.024e-23 / (1.381e-23)
B = 1.582e-26 / (1.381e-23)
r = np.linspace(-2, 2, 100)
epsilon = 2.4326066402741967e-07 #  = B^2 / 4A
r_0 = 0.34

k = (156 * B) / r_0 ** 14 - (42 * A) / r_0 ** 8
U = B / r**12 - A / r ** 6
V = 0.5 * k * (r - r_0) ** 2 + epsilon

plt.ylim(-5, 15)

plt.plot(r, U)
plt.plot(r, V)

plt.show()
```
