```python
# a)
import matplotlib.pyplot as plt
import numpy as np

A = 1.024e-23 / (1.381e-23)
B = 1.582e-26 / (1.381e-23)
r = np.linspace(0.3, 1, 1000)

U = B / r**12 - A / r ** 6
F = 12 * B / r ** 13 - 6 * A / r ** 7

line_1 = plt.plot(r, U, lw=2, label=r'U(r)')
plt.xlim(0.3, 0.8)
plt.ylim(-150, 100)

plt.twinx()
line_2 = plt.plot(r, F,'r', ls=':', lw=2.0, label=r'F(r)')
plt.xlim(0.3, 0.8)
plt.ylim(-1000, 1000)

lines = line_1 + line_2
labels = []
for line in lines:
    labels.append(line.get_label())
plt.legend(lines, labels)

plt.show()
```

```python
# b)
A = 1.024e-23 / (1.381e-23)
B = 1.582e-26 / (1.381e-23)
r = np.linspace(0.3, 1.0, 1000)
U = B / r**12 - A / r ** 6

plt.plot(r, U, 'k', lw=2., label='Lennard-Jones')

r0 = (2*B/A)**(1./6)
epsilon = B/r0**12 - A/r0**6
k = 156*B/r0**14 - 42*A/r0**8
V = 0.5 * k * (r-r0)**2 + epsilon

plt.plot(r, V, 'k', ls=':', lw=2., label='Harmonic approximation')

plt.xlim(0.3, 0.6)
plt.ylim(-120, 0)

plt.xlabel('r / nm')
plt.ylabel('Potential energy / K')

plt.legend(loc=4)
plt.show()
```
