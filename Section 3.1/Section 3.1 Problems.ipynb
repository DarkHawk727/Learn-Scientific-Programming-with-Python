```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-20, 20, 1000)
y1 = np.log(1 / np.cos(x) ** 2)
y2 = np.log(1 / np.sin(x) ** 2)

plt.xlabel('x')
plt.ylabel('y')

plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
# The functions are equal to 0 at x = nπ/2.
```

```python
# Look, I know that your not supposed / dont need to import things over and over....but it looks nice, and your the one coming to to ME for solutions, so deal with it.
import matplotlib.pyplot as plt
import numpy as np

K_m = 0.04
V_max = 0.1

S = np.linspace(-5, 5, 1000)
v = (V_max * S )/ (K_m + S)

plt.title('Michaelis-Menten Equation')
plt.xlabel('S')
plt.ylabel('v')

plt.plot(S, v)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.linspace(-10, 10, 1000)

sigma = 1
y1 = 1 / (sigma * math.sqrt(2 * math.pi)) * np.exp((-1 * x ** 2) / (2 * sigma ** 2))
sigma = 1.5
y2 = 1 / (sigma * math.sqrt(2 * math.pi)) * np.exp((-1 * x ** 2) / (2 * sigma ** 2))
sigma = 2
y3 = 1 / (sigma * math.sqrt(2 * math.pi)) * np.exp((-1 * x ** 2) / (2 * sigma ** 2))

plt.title('Normalized Gaussian Function')
plt.xlabel('x')
plt.ylabel('g(x)')

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```
