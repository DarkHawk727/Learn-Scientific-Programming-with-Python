```python
import matplotlib.pyplot as plt
import numpy as np
import math

k_1 = 0.3
k_2 = 0.1
A_0 = 0.2
t = np.linspace(-10, 10, 100)

A = A_0 * math.e ** (-1 * (k_1 + k_2) * t)
B = (k_1 / (k_1 + k_2)) * A_0 * (1 - math.e **((-1 * k_1 + k_2) * t))
C = (k_2 / (k_1 + k_2)) * A_0 * (1 - math.e **((-1 * k_1 + k_2) * t))

plt.plot(t, A)
plt.plot(t, B)
plt.plot(t, C)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy
from sympy import isprime

def isgaussprime(z: complex):
    a = int(z.real)
    b = int(z.imag)
    if a * b != 0:
        return isprime(a ** 2 + b ** 2)
    else:
        c = abs(a + b)
        return isprime(c) and c % 4 == 3

def connect(z1: complex, z2: complex):
    plt.plot([z1.real, z2.real], [z1.imag, z2.imag])
    
start = 5 + 23j
step = 1
z = start
next = None

while next != start:
    next = z + step
    connect(z, next)
    if isgaussprime(next):
        step *= 1j
    z = next

plt.axes() 

plt.show()
```
