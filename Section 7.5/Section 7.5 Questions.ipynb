```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

x, y = np.linspace(-10, 10, 500), np.linspace(-10, 10, 500)
X, Y = np.meshgrid(x, y)

f = np.sinc(np.hypot(X, Y))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, f,rstride=5, cstride=5, cmap=cm.coolwarm)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

dtype = np.dtype([("month", "i4"), ("day", "i4"), ("num_births", "i8")])
data = np.loadtxt(
    "C:/Users/Arjun Sarao/Learn-Scientific-Programming-with-Python-Solutions/LSPwP_accompanying_files/birthday-data.csv",
    dtype=dtype,
    delimiter=",",
    skiprows=1,
)

total_births = np.sum(data["num_births"])

heatmap = np.empty((12, 31))
heatmap[:] = np.nan

for month, day, births in data:
    heatmap[month - 1, day - 1] = births / total_births
fig = plt.figure()
ax = fig.add_subplot(111)
im = ax.imshow(heatmap, interpolation="nearest")
ax.set_yticks(range(12))
ax.set_yticklabels(
    ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
)
ax.set_title("Birth Probaility by Month and Day")
days = np.array(range(0, 31, 2))
ax.set_xticks(days)
ax.set_xticklabels(["{:d}".format(day + 1) for day in days])
ax.set_xlabel("Day of month")

cbar = fig.colorbar(ax=ax, mappable=im, orientation="horizontal")
cbar.set_label("Birthday Probability")

plt.show()
```
