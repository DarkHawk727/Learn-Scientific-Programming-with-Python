```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = x**3


fig, ax = plt.subplots(2, 3)

ax[0, 0].plot(x, y)
ax[0, 0].title.set_text('Normal')
ax[0, 1].plot(x, y)
ax[0, 1].title.set_text('x-Log')
ax[0, 1].set_xscale('log')
ax[0, 2].plot(x, y)
ax[0, 2].title.set_text('y-Log')
ax[0, 2].set_yscale('log')

ax[1, 0].plot(x, y)
ax[1, 0].title.set_text('Normal')
ax[1, 1].plot(x, y)
ax[1, 1].title.set_text('x-symLog')
ax[1, 1].set_xscale('symlog')
ax[1, 2].plot(x, y)
ax[1, 2].title.set_text('y-symLog')
ax[1, 2].set_yscale('symlog')

plt.subplots_adjust(wspace=0.7, hspace=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

text_file = "C:/Users/Arjun Sarao/Learn-Scientific-Programming-with-Python-Solutions/LSPwP_accompanying_files/moby-dick.txt"

letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
lcount = dict([(l, 0) for l in letters])

with open(text_file, 'r', encoding='utf-8') as f:
    for line in f:
        for letter in line:
            if letter in lcount:
                lcount[letter] += 1

# The total number of letters for normalization
norm = sum(lcount.values())
# A sorted list of (freq, letter) pairs
lfreqs = sorted([(v / norm * 100,k) for k, v in lcount.items()], reverse=True)

# Create a horizontal bar chart
plt.title("Frequency of letters in Moby Dick")
plt.xlabel("Letter")
plt.ylabel("Frequency")

plt.barh(range(len(lfreqs)), [freq for freq, letter in lfreqs], align='center')
plt.yticks(range(len(lfreqs)), [letter for freq, letter in lfreqs])

# Increase the space between the bars
plt.tick_params(axis='y', which='major', labelsize=7)

plt.show()
```
